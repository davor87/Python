# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
genero = ['hombre','mujer']
estado = ['bajos','medios','altos']
n=500
data_genero = []
data_estado = []

for i in range(n):
    data_genero.append(np.random.choice(genero))
    data_estado.append(np.random.choice(estado))
    
data_genero
data_estado[1:12]
```

```python
altura = 160 + 30 * np.random.randn(n)
peso = 65 + 25 * np.random.randn(n)
edad = 30 + 12 * np.random.randn(n)
ingresos = 18000 + 3500 * np.random.randn(n)

data = pd.DataFrame(
    {
        'Genero': data_genero,
        'Estado económico': data_estado,
        'Altura': altura,
        'Peso': peso,
        'Edad': edad,
        'Ingresos': ingresos,
        
    }
)
```

```python
data.head()
```

# agrupacion de datos

```python
grupo_genero = data.groupby('Genero')
grupo_genero.groups
```

```python
for names, groups in grupo_genero:
    print(names)
    print(groups)
```

```python
grupo_genero.get_group('mujer')
```

```python
grupo_doble = data.groupby(['Genero','Estado económico'])
for name,groups in grupo_doble:
    print(name)
    print(groups)
```
